```python
import matplotlib.pyplot as plt

train_loss_list, test_loss_list = [], []
train_loss_bbox_list, test_loss_bbox_list = [], []
train_loss_ce_list, test_loss_ce_list = [], []
train_loss_giou_list, test_loss_giou_list = [], []

with open('/root/Deformable-DETR/exps/r50_deformable_detr_plus_iterative_bbox_refinement_plus_plus_two_stage/log.txt') as f:
    for line in f:
        if 'epoch' in line:
            # print(eval(line))
            # print(type(eval(line)))
            line = eval(line)
            train_loss = line['train_loss']
            test_loss = line['test_loss']
            train_loss_bbox = line['train_loss_bbox']
            test_loss_bbox = line['test_loss_bbox']
            train_loss_ce = line['train_loss_ce']
            test_loss_ce = line['test_loss_ce']
            train_loss_giou = line['train_loss_giou']
            test_loss_giou = line['test_loss_giou']

            train_loss_list.append(train_loss)
            test_loss_list.append(test_loss)
            train_loss_bbox_list.append(train_loss_bbox)
            test_loss_bbox_list.append(test_loss_bbox)
            train_loss_ce_list.append(train_loss_ce)
            test_loss_ce_list.append(test_loss_ce)
            train_loss_giou_list.append(train_loss_giou)
            test_loss_giou_list.append(test_loss_giou)
```

```python
epochs = range(1, len(train_loss_list) + 1)

# train_loss, test_loss 그래프
plt.plot(epochs, train_loss_list, 'b', label='Training loss')
plt.plot(epochs, test_loss_list, 'r', label='Validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python

# train_loss_bbox, test_loss_bbox 그래프
plt.plot(epochs, train_loss_bbox_list, 'b', label='Training loss')
plt.plot(epochs, test_loss_bbox_list, 'r', label='Validation loss')
plt.title('Training and Validation Bbox Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python

# train_loss_ce, test_loss_ce 그래프
plt.plot(epochs, train_loss_ce_list, 'b', label='Training loss')
plt.plot(epochs, test_loss_ce_list, 'r', label='Validation loss')
plt.title('Training and Validation CE Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

```

```python

# train_loss_giou, test_loss_giou 그래프
plt.plot(epochs, train_loss_giou_list, 'b', label='Training loss')
plt.plot(epochs, test_loss_giou_list, 'r', label='Validation loss')
plt.title('Training and Validation Giou Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
